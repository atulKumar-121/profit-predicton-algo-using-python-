```python
import pandas as pd
```

```python
df=pd.read_csv("E:\\machine_learning\\50_Startups.csv")
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
df.State=encoder.fit_transform(df.State)
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
oh=OneHotEncoder()
df2=pd.DataFrame(oh.fit_transform(df[['State']]).toarray())
```

```python
df2
```

```python
merged_data=pd.concat([df,df2],axis=1)
print(merged_data)
```

```python
merged_data.pop(2)
merged_data.pop('State')
merged_data=merged_data.rename(columns={0:'California',1:'Florida'})
print(merged_data)
```

```python
from matplotlib import pyplot as plt
```

```python
x_rearch=merged_data['R&D Spend'].values.reshape(-1,1)
y_profit=merged_data['Profit'].values.reshape(-1,1)
plt.title("R&D Spend VS Profit")
plt.xlabel(" Profit")
plt.ylabel("R&D Spend")
plt.scatter(x_rearch,y_profit,marker="*")
plt.show()
```

```python
x_Administration =merged_data['Administration'].values.reshape(-1,1)
plt.title("Administration  VS Profit")
plt.xlabel(" Profit")
plt.ylabel("Administration ")
plt.scatter(x_Administration ,y_profit,marker="*")
plt.show()
```

```python
x_Marketing =merged_data['Marketing Spend'].values.reshape(-1,1)
plt.title("Marketing Spend  VS Profit")
plt.xlabel(" Profit")
plt.ylabel("Marketing Spend ")
plt.scatter(x_Marketing ,y_profit,marker="*")
plt.show()
```

```python
x=merged_data.iloc[:,[0,1,2,4,5]].values
y=merged_data.iloc[:,3:4].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
model=LinearRegression()
from sklearn.tree import DecisionTreeRegressor
model2=DecisionTreeRegressor()
model.fit(x_train,y_train)
model2.fit(x_train,y_train)
print("Model Coffeicent Using linear Regression",model.coef_)
print(" Model Intercept Using linear Regression",model.intercept_)
```

```python
y_predict=model.predict(x_test)
```

```python
print(y_predict)
```

```python
y_test
```

```python
y_predict2=model2.predict(x_test)
```

```python
print(y_predict2)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
print(" Model 1 Mean Absolute error", mean_absolute_error(y_test,y_predict))
print(" Model 2 Mean Absolute error", mean_absolute_error(y_test,y_predict2))
```

```python
print(" Model 1 Mean Squared error", mean_squared_error(y_test,y_predict))
print(" Model 2 Mean Squared error", mean_squared_error(y_test,y_predict2))
```
